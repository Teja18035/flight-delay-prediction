```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('flightdata.csv')
```

```python
dataset
```

```python
dataset.isnull().any()
```

```python
dataset['ARR_DEL15'].value_counts()
```

```python
dataset.corr()
```

```python
dataset['DEP_TIME'].fillna(dataset['DEP_TIME'].mean(),inplace = True)
dataset['DEP_DELAY'].fillna(dataset['DEP_DELAY'].mean(),inplace = True)
dataset['ARR_TIME'].fillna(dataset['ARR_TIME'].mean(),inplace = True)
dataset['ARR_DELAY'].fillna(dataset['ARR_DELAY'].mean(),inplace = True)
dataset['ARR_DEL15'].fillna(dataset['ARR_DEL15'].mode()[0],inplace = True)
```

```python
dataset.isnull().any()
```

```python
x = dataset[['DEP_TIME', 'DEP_DELAY', 'ARR_TIME', 'ARR_DELAY']]
```

```python
x
```

```python
y = dataset[['ARR_DEL15']]
```

```python
y
```

```python
x = x.values
y = y.values
```

```python
x, y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 0)
dtc.fit(x_train, y_train)
```

```python
y_pred_dtc = dtc.predict(x_test)
y_pred_dtc
```

```python
x_test_1 = np.array([[1521, 1, 2048, 16]])
x_test_1 =ss.fit_transform(x_test_1)
y_predict_1 = dtc.predict(x_test_1)
y_predict_1
```

```python
x_test_2 = np.array([[1654, 29, 1909, 24]])
```

```python
x_test_2 =ss.fit_transform(x_test_2)
```

```python
y_predict_2 = dtc.predict(x_test_2)
```

```python
y_predict_2
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
acc_dtc = accuracy_score(y_test, y_pred_dtc)
acc_dtc
```

```python
from sklearn.metrics import confusion_matrix
cm_dtc = confusion_matrix(y_test, y_pred_dtc)
cm_dtc
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred_lr = lr.predict(x_test)
y_pred_lr
```
